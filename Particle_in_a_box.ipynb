```python
import numpy as np
import matplotlib.pyplot as plt
from typing import Union

class ParticleInBox:
  """
  Class with static methods for study of the particle in the box energy levels
  """

  @staticmethod
  def energy(n: int, d: Union[int, float]) -> float:
    """
    Returns the energy of the particle in the box
    for the level with quantum number n,
    in arbitrary units
    ---
    Parameters:
    n:  int
        quantum number of an energy level
    d:  float
        width of the box, arb. units
    """
    return n**2 / d**2

  @staticmethod
  def wavelength(n: int, d: Union[int, float]) -> float:
    """
    Returns the wavelength of the wave function
    of the particle in the box for the level 
    with quantum number n, in arbitrary units
    ---
    Parameters:
    n:  int
        quantum number of an energy level
    d:  float
        width of the box, arb. units
    """
    return d * 2 / n

  @staticmethod
  def plot_levels(n_max: int, d: float, plot_waves: bool = True) -> None:
    """
    Plots the graph for n_max energy levels
    of the particle in the box
    ---
    Parameters:
    n_max:  int
            the highest quantum number to plot
    d:      float
            width of the box, arb. units
    """
    x = np.linspace(0, d, 101)
    for n in range(1, n_max+1):
      if plot_waves:
        plt.plot(x, ParticleInBox.energy(n,d) + 1*np.sin(2 * np.pi * x / ParticleInBox.wavelength(n, d)))
      plt.axhline(y=ParticleInBox.energy(n,d), color="grey", lw="0.5")
    plt.xlim(0, d)
```

```python
ParticleInBox.plot_levels(3, 1)
```

```python
ParticleInBox.plot_levels(10, 1)
```

```python
ParticleInBox.plot_levels(10, 0.1)
```

```python
ParticleInBox.plot_levels(10, 10)
```

```python
ParticleInBox.plot_levels(1000, 100, plot_waves=False)
```
